```python
import pandas as pd

#Task 1

visit_log = pd.read_csv("visit_log.csv", sep = '\;', engine='python')

#print(visit_log.head(5))

visit_log1 = visit_log.iloc[:50, 3:6]
#print(str(visit_log1.traffic_source))

visit_log1["source_type"] = visit_log1.traffic_source
visit_log1.loc[(visit_log1.traffic_source == 'google') | (visit_log1.traffic_source == 'yandex'), "source_type" ] = "organic"
visit_log1.loc[((visit_log1.traffic_source == 'paid') | (visit_log1.traffic_source == 'email')) & (visit_log1.region == 'Russia'), "source_type" ] = "ad"
visit_log1.loc[((visit_log1.traffic_source == 'paid') | (visit_log1.traffic_source == 'email')) & (visit_log1.region != 'Russia'), "source_type" ] = "other"

print(visit_log1)
```
