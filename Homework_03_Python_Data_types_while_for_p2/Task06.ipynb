```python
#Домашнее задание к лекции «Управляющие конструкции и коллекции» часть 2
#Task #6

cook_book = {
'салат': [
{'ingridient_name': 'сыр', 'quantity': 50, 'measure': 'гр'},
{'ingridient_name': 'томаты', 'quantity': 2, 'measure': 'шт'},
{'ingridient_name': 'огурцы', 'quantity': 20, 'measure': 'гр'},
{'ingridient_name': 'маслины', 'quantity': 10, 'measure': 'гр'},
{'ingridient_name': 'оливковое масло', 'quantity': 20, 'measure': 'мл'},
{'ingridient_name': 'салат', 'quantity': 10, 'measure': 'гр'},
{'ingridient_name': 'перец', 'quantity': 20, 'measure': 'гр'}
],
'пицца': [
{'ingridient_name': 'сыр', 'quantity': 20, 'measure': 'гр'},
{'ingridient_name': 'колбаса', 'quantity': 30, 'measure': 'гр'},
{'ingridient_name': 'бекон', 'quantity': 30, 'measure': 'гр'},
{'ingridient_name': 'оливки', 'quantity': 10, 'measure': 'гр'},
{'ingridient_name': 'томаты', 'quantity': 20, 'measure': 'гр'},
{'ingridient_name': 'тесто', 'quantity': 100, 'measure': 'гр'},
],
'лимонад': [
{'ingridient_name': 'лимон', 'quantity': 1, 'measure': 'шт'},
{'ingridient_name': 'вода', 'quantity': 200, 'measure': 'мл'},
{'ingridient_name': 'сахар', 'quantity': 10, 'measure': 'гр'},
{'ingridient_name': 'лайм', 'quantity': 20, 'measure': 'гр'},
]
}

portions_number = int(input("Введите количество порций: \n"))
#portions_number = 3

#--------- splitting to a list of keys and list of values 
keys_ = []
values_ = []
for key,value in cook_book.items():
    keys_.append(key)
    values_.append(value)


#-----------converting to simple list of dictionaries
res = []
for myList in values_:
    for item in myList:
        res.append(item)
        #print(item)
#print(res)

#------------------------converting to list of lists containing only dict values
res_1 = []
for i in range (0, len(res)):
    valuesss = res[i].values()
    res_1.append(list(valuesss))

#print(res_1)
#now we can work with res_1

product_name = []
product_weight = []
product_units = []
for x in range(0, len(res_1)):
    if x == 0:
        product_name.append(res_1[x][0])
        product_weight.append(res_1[x][1])
        product_units.append(res_1[x][2])
    else:
        if res_1[x][0] not in product_name:
            product_name.append(res_1[x][0])
            product_weight.append(res_1[x][1])
            product_units.append(res_1[x][2])
        else:
            val = res_1[x][0]
            index1 = product_name.index(val)
            product_weight[index1] = product_weight[index1] + res_1[x][1]
            #break

for x in range(0, len(product_name)):
    print(product_name[x],' ', int(product_weight[x]) * portions_number,' ',product_units[x])
```
