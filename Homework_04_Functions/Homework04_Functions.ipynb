```python

#Домашнее задание к лекции «Функции»

#Task #1. Пользователь по команде “p” может узнать владельца документа по его номеру
#Task #2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится
#Task #3. Пользователь по команде “l” может увидеть полную информацию по всем документам
#Task #4. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)
#Task #5.1 (additional). Пользователь по команде “ad” может добавить новый документ в данные
#Task #5.2 (additional). Пользователь по команде “d” может удалить документ из данных
#Task #5.3 (additional). Пользователь по команде “m” может переместить документ с полки на полку

documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'},
{'type': 'passport', 'number': '555', 'name': 'Иван Иванов'},
{'type': 'multipass', 'number': '7878', 'name': 'Лидия Никонова'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': [],
'5': ['555', '7878']
}


#------- command 'p', shows name of the owner based on doc number
def getDocOwner(doc_list):
    doc_number = str(input("Введите номер документа: \n "))
    list_key = [elem['number'] for elem in doc_list]
    if doc_number in list_key:
        for elements in doc_list:
            if elements['number'] == doc_number:
                print(f"Владелец документа {doc_number}: ", elements['name']," \n")
    else:
        print(f"Документ {doc_number} не найден в базе")


#------- command 's', based on doc number shows shelf number where the doc is kept
def getShelfNumber(shelfs_content):
    doc_number = str(input("Введите номер документа: \n "))
    list_val = [v for val in shelfs_content.values() for v in val]
    if doc_number in list_val:
        for key, val in shelfs_content.items():
            for x in val:
                if x == doc_number:
                    shelf_number = key
                    print(f"Документ {doc_number} хранится на полке ", shelf_number, " \n")
    else:
        print(f"Документ {doc_number} не найден в базе \n")


#------- command 'l', shows all info for all documents
def allInfoOutput(doc_list1, shelfs_content1):
    for elements in doc_list1:
        print(f"Документ # {elements['number']}, тип документа {elements['type']}, имя владельца {elements['name']}, хранится на полке", end = " ")
        for key, val in shelfs_content1.items():
            for x in val:
                if x == elements['number']:
                    shelf_number = key
                    print(f" номер {shelf_number}")
                    break
    print(" \n")


#-------------- command dir shows a list of existing shelfs and documents kept on them
def shelfInfoOutput(shelfs_content1):

    for key, val in shelfs_content1.items():
        print(f"Полка {key} содержит следующие документы ", val)
    print(" \n")


#------- command 'as', adding new shelf with the specified number
def addNewShelf(shelfs_content):
    filtered = {}
    filtered.update(shelfs_content)
    new_shelf_number = str(input("Введите номер добавляемой полки: \n "))
    if new_shelf_number not in filtered:
        filtered[new_shelf_number] = []
        print(f"Полка с номером {new_shelf_number} была добавлена. Текущий перечень полок:", end = " ")
    else:
        print(f"Полка с номером {new_shelf_number} уже существует. Текущий перечень полок:", end = " ")
    shelfs_content.clear()
    shelfs_content.update(sorted(filtered.items()))

    result = " ".join(str(key) for key in shelfs_content.keys())
    print(result)
    print(" \n")
    #print(shelfs_content)


#------- command 'ds', deleting the specified shelf
def deleteShelf(shelfs_content):
    del_shelf_number = str(input("Введите номер удаляемой полки: \n "))
    filtered = {}
    if del_shelf_number in shelfs_content:
        for key, val in shelfs_content.items():
            if key == del_shelf_number:
                if val == []:
                    print(f"Полка {del_shelf_number} удалена. Текущий перечень полок:", end=" ")
                else:
                    print(f"На полке {del_shelf_number} есть документы, удалите их перед удалением полки. Текущий перечень полок: ", end=" ")
                    filtered.update({key: val})
            else:
                filtered.update({key : val})
        shelfs_content.clear()
        shelfs_content.update(filtered)
    else:
        print(f"Полки {del_shelf_number} не существует. Текущий перечень полок: ", end=" ")
    result = " ".join(str(key) for key in shelfs_content.keys())
    print(result)
    print(" \n")
    #print(shelfs_content)


# ------- command 'ad', adding new document to the list
# ------- solution #1 for task for command 'ad', adding new document to the list

def addNewDocument(doc_list, shelfs_content):
    new_doc_number = str(int(input("Введите номер документа: ")))
    doc_type = str(input("Введите тип документа: "))
    doc_owner = str(input("Введите владельца документа: "))
    shelf_number = str(input("Введите полку для хранения: "))
    # checking if the shelf exists
    if shelf_number in shelfs_content:
        # list to check presence of the specified document
        list_dict_values = [v for values in shelfs_content.values() for v in values]
        if new_doc_number in list_dict_values:
            print(f"Документ с номером {new_doc_number} уже существует и добавлен быть не может ")
            print("Текущий список документов: ")
            allInfoOutput(doc_list, shelfs_content)
        else:
            new_dict = {}
            new_dict['type'] = doc_type
            new_dict['number'] = new_doc_number
            new_dict['name'] = doc_owner
            #print(new_dict)
            doc_list.append(new_dict)
            for key in shelfs_content.keys():
                if key == shelf_number:
                    shelfs_content[key].append(new_doc_number)
            #print('new ', shelfs_content)
            print(f"Документ {new_doc_number} добавлен. Текущий список документов: ")
            allInfoOutput(doc_list, shelfs_content)
    else:
        print(f"Полки {shelf_number} не существует. Добавьте полку командой 'as'")
        print("Текущий список документов: ")
        allInfoOutput(doc_list, shelfs_content)


"""
#--------------- another solution for task for command 'ad', adding new document to the list
#--------------- the code checks for both presence of the spedicifed shelf and document

def addNewDocument(doc_list, shelfs_content):
    #forming a list of keys using "doc_list" dictionary to add new records
    list_key = []
    for elem in doc_list:
        for key in elem.keys():
            if key not in list_key:
                list_key.append(key)

    new_doc_number = str(int(input("Введите номер документа: ")))
    doc_type = str(input("Введите тип документа: "))
    doc_owner = str(input("Введите владельца документа: "))
    shelf_number = str(input("Введите полку для хранения: "))

    #forming a list of the entered info related to the new record
    list_val = []
    list_val.append(doc_type)
    list_val.append(new_doc_number)
    list_val.append(doc_owner)
    #print("list_val ", list_val)

    #checking if the shelf exists
    if shelf_number in shelfs_content:

        #list to check presence of the specified document
        list_dict_values = [v for values in shelfs_content.values() for v in values]

        if new_doc_number in list_dict_values:
            print(f"Документ с номером {new_doc_number} уже существует и добавлен быть не может ")
            print("Текущий список документов: ")
            allInfoOutput(doc_list, shelfs_content)
        else:
            new_dict = {}
            new_dict = dict(zip(list_key, list_val))
            doc_list.append(new_dict)
            for key in shelfs_content.keys():
                if key == shelf_number:
                    shelfs_content[key].append(new_doc_number)
            print("Документ добавлен. Текущий список документов: ")
            allInfoOutput(doc_list, shelfs_content)

    else:
        print("Такой полки не существует. Добавьте полку командой 'as'")
        print("Текущий список документов: ")
        allInfoOutput(doc_list, shelfs_content)

"""

# ---------------- command "d", removing the specified document
def removeDocument(doc_list, shelfs_content):
    doc_number = str(input("Введите номер удаляемого документа: \n "))
    filtered = []
    list_of_all_values = [value for elem in doc_list for value in elem.values()]
    if doc_number in list_of_all_values:
        for element in doc_list:
            if element['number'] == doc_number:
                print(f"Документ {doc_number} удален")
                print("Текущий список документов: ")
            else:
                filtered.append(element)
        doc_list.clear()
        doc_list = filtered.copy()
        # -----------------------------------------------------------------------
        filtered1 = {}
        filtered1 = {key: [a for a in val if a != doc_number] for key, val in shelfs_content.items()}
        shelfs_content.clear()
        shelfs_content.update(filtered1)
        allInfoOutput(doc_list, shelfs_content)
        # -------------------------------------------------------------------------
    else:
        print(f"Документ {doc_number} не найден в базе")
        print("Текущий список документов: ")
        allInfoOutput(doc_list, shelfs_content)
    for element in doc_list:
        print(element)
    print(shelfs_content)
    return doc_list


# ---------------- command "m", moving the document from one shelf to another

def movingDocumenttoAnotherShelf(doc_list, shelfs_content):
    #print(shelfs_content)
    doc_number = str(input("Введите номер перемещаемого документа: \n "))
    shelf_number1 = str(input("Введите номер полки, на которую документ должен быть перемещен: \n "))
    # first to check if the document exists
    # second to check if the shelf exists
    filtered = {}
    list_val = [v for val in shelfs_content.values() for v in val]
    #to check if the specified documents exists
    if doc_number in list_val:
        #to check if the shelf-to-move-to exists
        if shelf_number1 in shelfs_content:
            for key, val in shelfs_content.items():
                #if the shelf is not the shelf-to-move-to
                if key != shelf_number1:
                    #if the checked shelf is assigned []
                    if val == []:
                        filtered.update({key: []})
                    else:
                        list3 = []
                        for v in val:
                            #if the checked document is not the one that should be moved
                            if v != doc_number:
                                list3.append(v)
                                filtered.update({key: list3})
                            else:
                                #if the shelf includes onle the document to move then [] is assigned
                                if len(val) == 1:
                                    print(" In val there is one doc ", v)
                                    filtered.update({key: []})
                else:
                    #if the shelf is the shelf-to-move-to
                    list3 = []
                    for v in val:
                        list3.append(v)
                    list3.append(doc_number)
                    filtered.update({key: list3})

            shelfs_content.clear()
            shelfs_content.update(sorted(filtered.items()))
            print("Документ перемещен")
            print("Текущий перечень полок: ")
            result = " ".join(str(key) for key in shelfs_content.keys())
            print(result)
            print("Текущий список документов: ")
            allInfoOutput(doc_list, shelfs_content)

        else:
            print("Такой полки не существует ")
            print("Текущий перечень полок: ", end = " ")
            result = " ".join(str(key) for key in shelfs_content.keys())
            print(result)
    else:
        print("Документ не найден в базе ")
        print("Текущий список документов: ")
        allInfoOutput(doc_list, shelfs_content)


# ---------------- to receive user's command
def enteredCommand():
    print("Введите, пожалуйста, команду из списка, приведенного внизу:")
    print("Команда 'p' выводит владельца документа по номеру документа")
    print("Команда 's' выводит по номеру документа номер полки, где хранится")
    print("Команда 'l' выводит полную информацию по всем документам")
    print("Команда 'dir' выводит информацию по существующим полкам и их содержимому")
    print("Команда 'as' позволит добавить новую полку")
    print("Команда 'ds' позволит удалить одну из существующих полок")
    print("Команда 'ad' позволяет добавить новый документ в данные")
    print("Команда 'd' позволяет удалить документ из данных")
    print("Команда 'm' позволяет переместить документ с одной полки на другую")
    print("Команда 'q' прекращает выполнение программы")
    command1 = str(input("Выбранная команда: "))
    return command1


#number = '10006'
#number = '11-2'
#number = 1234

command_list = ['p','s','l','dir','as','ds','ad','d','m','q']
while True:
    command = enteredCommand()
    if command not in command_list:
        print("Введеная команда не включена в предоставленный список, и исполнена быть не может \n")
    else:
        if command == 'p':
            print("Команда 'p' выводит владельца документа по номеру документа")
            getDocOwner(documents)
        elif command == 's':
            print("Команда 's' выводит по номеру документа номер полки, где хранится")
            getShelfNumber(directories)
        elif command == 'l':
            print("Команда 'l' выводит полную информацию по всем документам")
            allInfoOutput(documents, directories)
        elif command == 'dir':
            print("Команда 'dir' выводит информацию по существующим полкам и их содержимому")
            shelfInfoOutput(directories)
        elif command == 'as':
            print("Команда 'as' позволит добавить новую полку")
            addNewShelf(directories)
            #print("directories updated", directories)
        elif command == 'ds':
            print("Команда 'ds' позволит удалить одну из существующих полок")
            deleteShelf(directories)
        elif command == 'ad':
            print("Команда 'ad' позволяет добавить новый документ в данные")
            addNewDocument(documents, directories)
        elif command == 'd':
            print("Команда 'd' позволяет удалить документ из данных")
            documents = removeDocument(documents, directories)
            #print("-------------------------")
            #print(documents)
            #print(directories)
        elif command == 'm':
            print("Команда 'm' позволяет переместить документ с одной полки на другую")
            movingDocumenttoAnotherShelf(documents, directories)         
        elif command == 'q':
            print("Команда 'q' прекращает выполнение программы")
            break

```
